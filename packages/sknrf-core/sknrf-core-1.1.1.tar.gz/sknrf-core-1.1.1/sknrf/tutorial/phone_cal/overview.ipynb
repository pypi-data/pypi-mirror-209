## Installation
Qualcom ETTS:
    - Copy to C:\ETTS_TCF
    - Ctrl-F for 'C:' text and add all C:\Withnail_Data and C:\Measured_Data directories
    - Jongsoo Code: Install MSMQ, Go to Control Panel -> Add Remove Program -> Add Windows Component and select “Message Queuing”.
