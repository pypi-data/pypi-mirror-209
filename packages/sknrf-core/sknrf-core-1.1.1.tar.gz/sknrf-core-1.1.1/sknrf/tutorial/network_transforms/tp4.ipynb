```python
from sympy import symbols, Eq, Matrix, Rational, solve, simplify, factor, init_printing
from sympy.functions import re, im, sqrt
from sympy.functions import conjugate as conj
init_printing()
```

```python
Z_0 = symbols("Z_0")
T_p_1_1, T_p_1_2, T_p_1_3, T_p_1_4 = symbols("T_p_1_1 T_p_1_2 T_p_1_3 T_p_1_4")
T_p_2_1, T_p_2_2, T_p_2_3, T_p_2_4 = symbols("T_p_2_1 T_p_2_2 T_p_2_3 T_p_2_4")
T_p_3_1, T_p_3_2, T_p_3_3, T_p_3_4 = symbols("T_p_3_1 T_p_3_2 T_p_3_3 T_p_3_4")
T_p_4_1, T_p_4_2, T_p_4_3, T_p_4_4 = symbols("T_p_4_1 T_p_4_2 T_p_4_3 T_p_4_4")
```

```python
#tp4t
Z_1, Z_2, Z_3, Z_4 = symbols("Z_1 Z_2 Z_3 Z_4")
Z_1_F = 1/(2*(re(Z_0)*re(Z_1))**(1/2))
Z_2_F = 1/(2*(re(Z_0)*re(Z_2))**(1/2))
Z_3_F = 1/(2*(re(Z_0)*re(Z_3))**(1/2))
Z_4_F = 1/(2*(re(Z_0)*re(Z_4))**(1/2))
Matrix_1 = Matrix([[Z_1_F*(Z_0 + conj(Z_1)), Z_1_F*(Z_0 - conj(Z_1))   , 0                      ,                       0],
                   [Z_1_F*(Z_0 - Z_1)      , Z_1_F*(Z_0 + Z_1)         , 0                      ,                       0],
                   [0                      , 0                         , Z_2_F*(Z_0 + conj(Z_2)), Z_2_F*(Z_0 - conj(Z_2))],
                   [0                      , 0                         , Z_2_F*(Z_0 - Z_2)      , Z_2_F*(Z_0 + Z_2)      ]])
Matrix_2 = Matrix([[T_p_1_1, T_p_1_2, T_p_1_3, T_p_1_4],
                   [T_p_2_1, T_p_2_2, T_p_2_3, T_p_2_4],
                   [T_p_3_1, T_p_3_2, T_p_3_3, T_p_3_4],
                   [T_p_4_1, T_p_4_2, T_p_4_3, T_p_4_4]])
Matrix_3 = Matrix([[Z_3_F*(Z_0 + conj(Z_3)), Z_3_F*(Z_0 - conj(Z_3))   , 0                      ,                       0],
                   [Z_3_F*(Z_0 - Z_3)      , Z_3_F*(Z_0 + Z_3)         , 0                      ,                       0],
                   [0                      , 0                         , Z_4_F*(Z_0 + conj(Z_4)), Z_4_F*(Z_0 - conj(Z_4))],
                   [0                      , 0                         , Z_4_F*(Z_0 - Z_4)      , Z_4_F*(Z_0 + Z_4)      ]])
T = simplify(factor(Matrix_1.inv())*factor(Matrix_2*Matrix_3))
T.T
```
