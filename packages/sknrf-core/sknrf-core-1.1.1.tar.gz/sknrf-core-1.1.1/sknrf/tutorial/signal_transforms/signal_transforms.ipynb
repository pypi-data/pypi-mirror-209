```python
# Matplotlib figures generated in the notebook inline with the rest of the script
%matplotlib inline

import numpy as np # Import NumPy for arrays and basic linear algebra
from collections import namedtuple
import matplotlib.pyplot as plt # Import PyPlot for basic plotting.
import matplotlib as mpl # Import Matplotlib for advanced plotting.
from mpl_toolkits.mplot3d import Axes3D # Import Matplotlib 3D Plots

from sknrf.model.settings import Settings
from sknrf.devices.signal import *
plt.style.use('bmh')
```

```python
Settings.f0 = 1e9
Settings.num_harmonics = 3
Settings.t_stop = 10e-6
Settings.t_step = 1e-6
envelope_signal = EnvelopeSignal(np.ones((Settings.f_points, Settings.t_points), dtype=complex))
envelope_signal[0,:] = 0
envelope_signal[1,:] = 2*np.arange(1, 2.0, 0.1)
envelope_signal[2,:] = 0
envelope_signal[3,:] = 1*np.arange(1, 2.0, 0.1)
print(envelope_signal)
```

```python
plot_signal = nd_to_2d_formatter(envelope_signal)
plt.plot(plot_signal.indep_map['time'], plot_signal)
```

```python
time_signal = ts(envelope_signal)
plot_signal = nd_to_2d_formatter(time_signal)
plt.plot(plot_signal.indep_map['time_c'], plot_signal)
```

```python
time_envelope_signal = tes(envelope_signal,te_stop=10e-6, te_points=7666)
plot_signal = nd_to_2d_formatter(time_envelope_signal)
plt.plot(plot_signal.indep_map['time_e'], plot_signal)
```

```python
frequency_signal = fs(envelope_signal)
plot_signal = nd_to_2d_formatter(frequency_signal)
plt.plot(plot_signal.indep_map['freq_m'], plot_signal)
```

```python
frequency_spectrum_signal = fss(envelope_signal)
plot_signal = nd_to_2d_formatter(frequency_spectrum_signal)
for index in range(plot_signal.shape[1]):
    plt.stem(plot_signal.indep_map['freq'], plot_signal[:, index], markerfmt='b^')
```

```python
power_signal = ps(envelope_signal)
plot_signal = nd_to_2d_formatter(power_signal)
plt.plot(np.array(plot_signal))
```
